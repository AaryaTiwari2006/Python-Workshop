### Create csv

```python
import pandas as pd
data={
    "Name":["Aarya","Moriom","Vanshika","Sayyanika","Ishita","Nishita","Shikha","Roushan","Supritam","Sayak"],
    "Height":[170,160,172,180,190,184,164,189,130,171]
}
df=pd.DataFrame(data)
df.to_csv("height.csv",index=False)
```

## Unideviate

```python
df=pd.read_csv("height.csv")
print(df)
mean=df["Height"].mean()
print("mean of students height:", mean)
```

### Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df["Height"], bins=5,
edgecolor='black')
plt.title("Histogram of student hight")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()
         
```

## for bivarite


```python
data={
    "Exercise":[10,15,20,25,30,35,40,45,50,55],
    "Heart rate":[80,85,88,90,92,95,97,100,102,105]
}
df=pd.DataFrame(data)
df.to_csv("heartrate.csv",index=False)
print(df)
```

```python
df=pd.read_csv("heartrate.csv")
print("Bivariate data:")
print(df)
plt.scatter(df["Exercise"],
df["Heart rate"], color='red',
marker='o')
plt.title("Exercise and heart rate graph")
plt.xlabel("Exercsie")
plt.ylabel("heart rate")
plt.show()
```
